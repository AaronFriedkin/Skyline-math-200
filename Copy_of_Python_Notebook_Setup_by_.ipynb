```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

Codebook 1

```python
#Manage data for variable 1
df['Smoke_Reg'] = df['H4TO3']
df['Smoke_Reg'] = df['Smoke_Reg'].loc[ df['Smoke_Reg'] !=' ' ]
df['Smoke_Reg'] = df['Smoke_Reg'].loc[ df['Smoke_Reg'] !='6' ]
df['Smoke_Reg'] = df['Smoke_Reg'].loc[ df['Smoke_Reg'] !='7' ]
df['Smoke_Reg'] = df['Smoke_Reg'].loc[ df['Smoke_Reg'] !='8' ]
df['Smoke_Reg'].value_counts()
```

```python
df['Smoke_Reg'].replace( ['0','1'] , ['NO','YES'], inplace=True)
df['Smoke_Reg'].value_counts()
```

```python
rp.summary_cat(df['Smoke_Reg'])
```

```python
#Graphs and tables or statistics
df['Smoke_Reg'].value_counts().plot(kind ='pie', autopct='%1.1f%%')
```

```python
#Manage data for variable 1
df['HEALTH_DIET'] = df['H4GH1']
```

```python
df['HEALTH_DIET'] = df['HEALTH_DIET'].loc[ df['HEALTH_DIET'] !=' ' ]
df['HEALTH_DIET'] = df['HEALTH_DIET'].loc[ df['HEALTH_DIET'] !='6' ]
df['HEALTH_DIET'] = df['HEALTH_DIET'].loc[ df['HEALTH_DIET'] !='7' ]
df['HEALTH_DIET'] = df['HEALTH_DIET'].loc[ df['HEALTH_DIET'] !='8' ]
df['HEALTH_DIET'].value_counts()
```

```python
df['HEALTH_DIET'].replace(['2','1','3','4','5'],['Excellent','Very Good','Good','Fair','Poor'],inplace=True)
```

```python
df['HEALTH_DIET'].value_counts()
```

```python
#Graphs and tables or statistics
df['HEALTH_DIET'].value_counts().plot(kind ='pie', autopct='%1.1f%%')
```

```python
pd.crosstab( df['HEALTH_DIET'], df['Smoke_Reg'], normalize='index').plot(kind = 'bar' , stacked = True)
```
