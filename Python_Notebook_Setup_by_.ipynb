```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Smoked_Cigs'] = df['H4TO11']
```

```python
rp.summary_cat(df['Smoked_Cigs'])
```

```python

#Manage data for variable 1

df['Smoked_Cigs'] = df['Smoked_Cigs'].loc[ df['Smoked_Cigs'] !='2' ]
df['Smoked_Cigs'] = df['Smoked_Cigs'].loc[ df['Smoked_Cigs'] !='3' ]
df['Smoked_Cigs'] = df['Smoked_Cigs'].loc[ df['Smoked_Cigs'] !='4' ]
df['Smoked_Cigs'] = df['Smoked_Cigs'].loc[ df['Smoked_Cigs'] !='1' ]
df['Smoked_Cigs'] = df['Smoked_Cigs'].loc[ df['Smoked_Cigs'] !='6' ]
df['Smoked_Cigs'] = df['Smoked_Cigs'].loc[ df['Smoked_Cigs'] !='8' ]
df['Smoked_Cigs'].value_counts()
```

```python
df['Smoked_Cigs'].replace( ['7',' '] , ['NO','YES'], inplace=True)
df['Smoked_Cigs'].value_counts()
```

```python

#Graphs and tables or statistics
df['Smoked_Cigs'].value_counts().plot(kind ='pie', autopct='%1.1f%%')
```

```python
#Manage data for variable 1
df['Been_Arrested'] = df['H4CJ1']
```

```python
rp.summary_cat(df['Been_Arrested'])
```

```python
df['Been_Arrested'] = df['Been_Arrested'].loc[ df['Been_Arrested'] !='2' ]
df['Been_Arrested'] = df['Been_Arrested'].loc[ df['Been_Arrested'] !='3' ]
df['Been_Arrested'] = df['Been_Arrested'].loc[ df['Been_Arrested'] !='4' ]
df['Been_Arrested'] = df['Been_Arrested'].loc[ df['Been_Arrested'] !='9' ]
df['Been_Arrested'] = df['Been_Arrested'].loc[ df['Been_Arrested'] !='6' ]
df['Been_Arrested'] = df['Been_Arrested'].loc[ df['Been_Arrested'] !='7' ]
df['Been_Arrested'].value_counts()
```

```python
df['Been_Arrested'].replace( ['0',' '] , ['NO','YES'], inplace=True)
df['Been_Arrested'].value_counts()
```

```python
#Graphs and tables or statistics
df['Been_Arrested'].value_counts().plot(kind ='pie', autopct='%1.1f%%')
```

```python
pd.crosstab( df['Been_Arrested'], df['Smoked_Cigs'], normalize='index').plot(kind = 'bar' , stacked = True)
```
